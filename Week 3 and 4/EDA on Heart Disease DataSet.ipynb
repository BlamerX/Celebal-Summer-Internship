```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./heart-disease.csv')
df.head(10)
```

```python
df['age'].plot.hist(bins=10);
```

```python
sex_counts = df['sex'].value_counts()

labels = ['Female', 'Male'] 
plt.pie(sex_counts, labels=labels, autopct='%1.1f%%')
plt.title('Sex Distribution')
plt.axis('equal')
plt.show()
```

```python
over_50 = df[df["age"] > 50]
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") # This changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Customizing the x and y axis limitations

plt.style.use("ggplot")

# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                        ncols=1,
                        figsize=(10,10),sharex=True)

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Change the x limitation
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax0
ax0.axhline(over_50["chol"].mean(),linestyle='-.');

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Change the x limitation
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle='-.')

# Add a title toi the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
